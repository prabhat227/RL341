```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)


num_points = 50
x = np.random.rand(num_points)
print(x)
y = np.random.rand(num_points)
print(y)


def create_scatter_plot(x, y, title):
    plt.scatter(x, y)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(title)
    plt.grid(True)
    plt.show()

y_positive = x + 0.2 * np.random.randn(num_points)
create_scatter_plot(x, y_positive, "Positive Correlation")


y_negative = -x + 0.2 * np.random.randn(num_points)
create_scatter_plot(x, y_negative, "Negative Correlation")


y_zero = 0.2 * np.random.randn(num_points)
create_scatter_plot(x, y_zero, "Zero Correlation")


```

```python
import seaborn as sns
data = np.vstack((x, y_positive, y_negative, y_zero))
labels = ["X", "Y Positive", "Y Negative", "Y Zero"]
correlation_matrix = np.corrcoef(data)
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", xticklabels=labels, yticklabels=labels)
plt.title("Correlation Heatmap")
plt.show()
```
